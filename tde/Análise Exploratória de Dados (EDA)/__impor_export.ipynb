```python
# Colocando o mês como primeira coluna e somando o numero de linhas e colunas
# Para usar uma coluna expecifica podemos usar (...).sum()['nomeDaColuna'])

pd.DataFrame(arquivo1.groupby(["MÊS", "ANO"]).sum())
```

```python
# Verifica os valores nulos de cada coluna.

print('Colunas:')
print(arquivo1.isnull().sum())
print()

# Número de linhas e colunas

num_lin, num_cols = arquivo1.shape
print(f"O conjunto de dados tem {num_lin} linhas e {num_cols} colunas.")
print()

# Nomes de cada coluna

column_names =arquivo1.columns.tolist()
print(f"Os nomes das colunas são: {column_names}")
print()

# Obter a média, mediana e devio do padrão

med_valor = arquivo1['IMPORTADO / EXPORTADO'].mean()
mediana_valor = arquivo1['IMPORTADO / EXPORTADO'].median()
desvio = arquivo1['IMPORTADO / EXPORTADO'].std()
print('A média é {}, a mediana é {} e o desvio de padrão é {}' .format(med_valor, mediana_valor, desvio))

# Função que agrupa os dados em uma coluna para obter estatísticas 

group_data = arquivo1.groupby('IMPORTADO / EXPORTADO').describe()
print(group_data)
```

```python
# Pegando informação expecifica do arquivo

df_export = arquivo1[(arquivo1['OPERAÇÃO COMERCIAL'] == 'EXPORTAÇÃO') & (arquivo1['ANO'] == 2022)]

# Plota o gráfico de barras
plt.bar(df_export['MÊS'], df_export['DISPÊNDIO / RECEITA'])
plt.xlabel('Mês')
plt.ylabel('Quantidade Exportada (1000 m³)')
plt.title('Exportação no ano 2022')
plt.show()
```

```python
# remove o ano 2023
df_filtered = arquivo1[arquivo1['ANO'] < 2023]

# agrupa os dados por ano e soma as produções
producao_por_ano = df_filtered.groupby('ANO')['DISPÊNDIO / RECEITA'].sum()

# ordena o resultado do agrupamento pela coluna de produção em ordem decrescente e pega o primeiro valor, correspondente ao ano com maior produção.
ano_mais_produtivo = producao_por_ano.sort_values(ascending=False).index[0]

# ordena o resultado do agrupamento pela coluna de produção em ordem decrescente e pega o primeiro valor, correspondente ao ano com maior produção.
ano_menos_produtivo = producao_por_ano.sort_values(ascending=True).index[0]

print(f"O ano mais produtivo foi {ano_mais_produtivo} e o ano menos produtivo foi {ano_menos_produtivo}.")

# encontra o índice da maior e menor produção
idx_maior = producao_por_ano.argmax()
idx_menor = producao_por_ano.argmin()

# define as cores das barras
cores = ['r' if i == idx_menor else 'g' if i == idx_maior else 'b' for i in range(len(producao_por_ano))]

# plota o gráfico de barras
producao_por_ano.plot(kind='bar', color=cores, figsize=(10,5))

# define os títulos do gráfico e dos eixos
plt.title('Importação e exportação por Ano')
plt.xlabel('Ano')
plt.ylabel('Produção (1000m³)')

# exibe o gráfico
plt.show()
```
