```python
# Leitura do arquivo
import pandas as pd

arquivo1 = './Dados/importacoes-exportacoes-petroleo-2000-2023.xls'
arquivo2 = 'https://raw.githubusercontent.com/GabrielQuinteiro/TDE-3sem/main/producao-gas-combustivel-1000m3-2000-2023.xls'

df1 = pd.read_excel(arquivo1)
df2 = pd.read_excel(arquivo2)
# Imprimindo arquivo completo
```

```python
df1["ANOMES"] = df1["ANO"].astype(str)+'-'+df1["MÊS"]
df1 = df1.drop(columns=["ANO","MÊS"]).set_index("ANOMES")
```

```python
#Exportação é marcada como negativa (sai do país)
df1.loc[df1["OPERAÇÃO COMERCIAL"] == "EXPORTAÇÃO", "IMPORTADO / EXPORTADO"] = df1[df1["OPERAÇÃO COMERCIAL"] == "EXPORTAÇÃO"]["IMPORTADO / EXPORTADO"].apply(lambda x: x *-1)
df1["OPERAÇÃO COMERCIAL"] = df1["OPERAÇÃO COMERCIAL"].astype('category')
print(df1["OPERAÇÃO COMERCIAL"])
df1["OPERAÇÃO COMERCIAL"]  = df1["OPERAÇÃO COMERCIAL"].cat.codes
#df1 = df1.drop(columns=["OPERAÇÃO COMERCIAL"])
```

```python
#Total produção nacional por ano
df2["ANOMES"] = df2["ANO"].astype(str)+'-'+df2["MÊS"]
df2 = df2.drop(columns=["ANO"]).groupby(["ANOMES"], group_keys=False).sum()
df2
```
