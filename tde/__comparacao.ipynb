```python
# pacote de processamento dos dados
import pandas as pd

# pacote de visualização de dados
import matplotlib.pyplot as plt
```

```python
# lendo os arquivos

df1 = pd.read_excel('./Dados/importacoes-exportacoes-petroleo-2000-2023.xls')
df2 = pd.read_excel('https://raw.githubusercontent.com/GabrielQuinteiro/TDE-3sem/main/producao-gas-combustivel-1000m3-2000-2023.xls')
```

```python
#

plt.plot(df1['ANO'], df1['DISPÊNDIO / RECEITA'], label='Dataset 1')
plt.plot(df2['ANO'], df2['PRODUÇÃO'], label='Dataset 2')
plt.legend()
plt.xlabel('Data')
plt.ylabel('Valor')
plt.title('Comparação dos datasets 1 e 2')
plt.show()
```

```python
plt.scatter(df1['MÊS'], df1['IMPORTADO / EXPORTADO'], label='Dataset 1')
plt.scatter(df2['MÊS'], df2['PRODUÇÃO'], label='Dataset 2')
plt.legend()
plt.title('Comparação da relação entre x e y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
